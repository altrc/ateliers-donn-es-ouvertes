```python
import pandas as pd
```

```python
sc = pd.read_csv ('Documents/atelierdonneesouvertes/liste-financement-frqsc-2018-2019.csv', usecols = ['Etablissement', 'Domaine_de_recherche', 'Montant_total', ])
```

```python
sc
```

```python
dict_etablissements_sc=set(sc['Etablissement'].tolist())
```

```python
dict_etablissements_sc
```

```python
liste_montantsParEtablissements_sc=[]
for et in dict_etablissements_sc:
    montants_sc=sc.loc[sc['Etablissement']== et]    
    dict_etablissements_sc = {'org': et, 'total': sum(montants_sc['Montant_total'].tolist())}
    liste_montantsParEtablissements_sc.append(dict_etablissements_sc)
```

```python
liste_montantsParEtablissements_sc
```

```python
sortedByTotalByEtablissement_sc = sorted(liste_montantsParEtablissements_sc, key=lambda d: d['total'], reverse=True)
```

```python
sortedByTotalByEtablissement_sc
```

```python
dict_domaines_sc=set(sc['Domaine_de_recherche'].tolist())
```

```python
dict_domaines_sc
```

```python
liste_montantsDomaine_sc=[]
for et in dict_domaines_sc:
    montants_sc=sc.loc[sc['Domaine_de_recherche']== et]    
    dict_domaines_sc = {'domaine': et, 'total': sum(montants_sc['Montant_total'].tolist())}
    liste_montantsDomaine_sc.append(dict_domaines_sc)
```

```python
liste_montantsDomaine_sc
```

```python
sortedByTotalByDomaine_sc = sorted(liste_montantsDomaine_sc, key=lambda d: d['total'], reverse=True)
```

```python
sortedByTotalByDomaine_sc
```

```python
montantTotalsc = sum(item['total'] for item in liste_montantsParEtablissements_sc)
```
