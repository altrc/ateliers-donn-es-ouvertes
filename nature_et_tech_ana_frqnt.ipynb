```python
import pandas as pd
```

```python
## Obtenir les colonnes Établissementent, Domaine de recherche et Montants total dans le domaine de la nature et technologies
nt = pd.read_csv('liste-financement-frqnt-2018-2019.csv', usecols = ['Etablissement', 'Domaine_de_recherche', 'Montant_total', ])
```

```python
nt
```

```python
## Créer un dictionnaire de tous les établissements sans doublon
dict_etablissements_nt=set(nt['Etablissement'].tolist())
```

```python
dict_etablissements_nt
```

```python
## Créer une nouvelle liste avec un loop pour obtenir montant total à chaque établissement
liste_montantsParEtablissements_nt=[]
for et in dict_etablissements_nt:
    montants_nt=nt.loc[nt['Etablissement']== et]    
    dict_etablissements_nt = {'org': et, 'total': sum(montants_nt['Montant_total'].tolist())}
    liste_montantsParEtablissements_nt.append(dict_etablissements_nt)
```

```python
liste_montantsParEtablissements_nt
```

```python
sortedByTotalByEtablissement_nt = sorted(liste_montantsParEtablissements_nt, key=lambda d: d['total'], reverse=True)
```

```python
## du plus petit au plus grands montants par établissement
sortedByTotalByEtablissement_nt
```

```python
## Créer un dictionnaire de tous les domaines sans doublon
dict_domaines_nt=set(nt['Domaine_de_recherche'].tolist())
```

```python
dict_domaines_nt
```

```python
## Créer une nouvelle liste avec un loop pour obtenir montant total à chaque établissement
liste_montantsDomaine_nt=[]
for et in dict_domaines_nt:
    montants_nt=nt.loc[nt['Domaine_de_recherche']== et]    
    dict_domaines_nt = {'domaine': et, 'total': sum(montants_nt['Montant_total'].tolist())}
    liste_montantsDomaine_nt.append(dict_domaines_nt)
```

```python
liste_montantsDomaine_nt
```

```python
sortedByTotalByDomaine_nt = sorted(liste_montantsDomaine_nt, key=lambda d: d['total'], reverse=True)
```

```python
## du plus petit au plus grands montants par domaine
sortedByTotalByDomaine_nt
```

```python
montantTotalNT = sum(item['total'] for item in liste_montantsParEtablissements_nt)
```
