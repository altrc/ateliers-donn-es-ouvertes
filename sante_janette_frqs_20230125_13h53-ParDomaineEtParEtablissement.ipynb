```python
import pandas as pd
```

```python
## Obtenir les colonnes Établissement et Montants total dans le domaine de la santé
sante = pd.read_csv('liste-financement-frqs-2018-2019.csv', usecols = ['Etablissement', 'Domaine_de_recherche', 'Montant_total', ])
```

```python
print(sante)
```

```python
## Créer un dictionnaire de tous les établissements sans doublon
dict_etablissements=set(sante['Etablissement'].tolist())
```

```python
dict_etablissements
```

```python
## Créer une nouvelle liste avec un loop pour obtenir montant total à chaque établissement
liste_montantsParEtablissements=[]
for et in dict_etablissements:
    montants=sante.loc[sante['Etablissement']== et]    
    dict_etablissements = {'org': et, 'total': sum(montants['Montant_total'].tolist())}
    liste_montantsParEtablissements.append(dict_etablissements)
```

```python
liste_montantsParEtablissements
```

```python
sortedByTotalByEtablissement = sorted(liste_montantsParEtablissements, key=lambda d: d['total'])
```

```python
## du plus petit au plus grands montants par établissement
sortedByTotalByEtablissement
```

```python
## Créer un dictionnaire de tous les domaines sans doublon
dict_domaines=set(sante['Domaine_de_recherche'].tolist())
```

```python
dict_domaines
```

```python
## Créer une nouvelle liste avec un loop pour obtenir montant total à chaque établissement
liste_montantsDomaine=[]
for et in dict_domaines:
    montants=sante.loc[sante['Domaine_de_recherche']== et]    
    dict_domaines = {'domaine': et, 'total': sum(montants['Montant_total'].tolist())}
    liste_montantsDomaine.append(dict_domaines)
```

```python
liste_montantsDomaine
```

```python
sortedByTotalByDomaine = sorted(liste_montantsDomaine, key=lambda d: d['total'])
```

```python
## du plus petit au plus grands montants par domaine
sortedByTotalByDomaine
```

```python
montantTotalSante = sum(item['total'] for item in liste_montantsParEtablissements)
```
